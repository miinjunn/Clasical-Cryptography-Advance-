## **Enkripsi** <br>

>Ekspektasi: <br>
Plaintext 	: we are discovered flee at once <br>
Key 		: tomato <br>

> Ciphertext 	: ROFOM ACDTL EDSEE EACOQ WEIVR LENEC 

```python
import numpy as np, string, random
from collections import Counter


plainText = input("Plaintext \t: ").upper().replace(" ", "")
key = input("Key \t\t: ")
key = list(key.upper())
sortkey = []
sortkey.append(key)
sortkey.append([])
i = 0
while i<len(key):
  sortkey[1].append(i+1)
  i+=1
sortkey = np.array(sortkey)
newsortkey = sortkey [ :, sortkey[0].argsort()]
sortkey = newsortkey.tolist()
sortkey.append([])
hitung = Counter(sortkey[0])
i = 0
n = 1
while i<len(key):
  if hitung[sortkey[0][i]]>1:
    for j in range(hitung[sortkey[0][i]]):
        sortkey[2].append(n)
    i+=(hitung[sortkey[0][i]]-1)
  else:
      sortkey[2].append(n)
  n+=1
  i+=1
sortkey = np.array(sortkey)
newsortkey = sortkey [ :, sortkey[1].argsort()]

print("\nTabel dan Urutan"+"\nBerdasarkan Key\t:")
print("  "+np.array2string(newsortkey[2],formatter={'str_kind': lambda x: x},separator=' ')[1:-1])
enkripsi = []
enkripsi.append(key)
for i in range (0, len(plainText), len(key)):
  x = plainText[i:i+len(key)]
  x = list(x)
  enkripsi.append(x)
while len(enkripsi[len(enkripsi)-1])<len(key):
  enkripsi[len(enkripsi)-1].append(random.choice(string.ascii_uppercase))
  
  # enkripsi[len(enkripsi)-1].append("X")
enkripsi = np.array(enkripsi)
print(" "+np.array2string(enkripsi[1:len(enkripsi)],formatter={'str_kind': lambda x: x},separator=' ')[1:-1])
newenkripsi = enkripsi [ :, enkripsi[0].argsort()]
hitung = Counter(newenkripsi[0])
hasil = []
i = 0
while i < len(newenkripsi[0]):
  if hitung[newenkripsi[0][i]] == 1:
    for j in range(1, len(newenkripsi)):
      hasil.append(newenkripsi[j][i])
  else:
    for j in range(1, len(newenkripsi)):
      for k in range(i, hitung[newenkripsi[0][i]] + i):
        hasil.append(newenkripsi[j][k])
    i += (hitung[newenkripsi[0][i]] - 1)
  i += 1
print("\nCiphertext \t:", end=' ')
for i in range(len(hasil)):
  print(hasil[i], end='')
  if (i + 1) % (len(newenkripsi) - 1) == 0:
    print(" ", end='')
```
